# **Variables estadísticas útiles**

## Media ($\bar{x}$)

La media, denotada como \(\bar{x}\), se calcula como la suma de todos los valores dividida por el número de observaciones. Es una medida de tendencia central que representa el valor promedio de un conjunto de datos.

La fórmula para la media es:

$\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i$

Donde:
- $\bar{x}$ es la media,
- $n$ es el número de observaciones,
- $x_i$ son los valores individuales del conjunto de datos.

## Desviación Estándar (\($\sigma$\))

La desviación estándar, denotada como \(\sigma\), **mide la dispersión de los datos con respecto a la media**. Es una medida de la variabilidad en un conjunto de datos.

La fórmula para la desviación estándar es:

$\sigma = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2}$

Donde:
- \(\sigma\) es la desviación estándar,
- \(\bar{x}\) es la media,
- \(n\) es el número de observaciones,
- \(x_i\) son los valores individuales del conjunto de datos.



```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios para la distribución
np.random.seed(42)
datos = np.random.normal(loc=10, scale=5, size=1000)

# Calcular métricas estadísticas
media = np.mean(datos)
mediana = np.median(datos)
desviacion_estandar = np.std(datos)

# Crear histograma
plt.hist(datos, bins=30, alpha=0.7, color='blue', label='Distribución de Datos')

# Líneas verticales para métricas estadísticas
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label=f'Media: {media:.2f}')
plt.axvline(mediana, color='green', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana:.2f}')
plt.axvline(media + desviacion_estandar, color='orange', linestyle='dashed', linewidth=2, label='Desviación Estándar')
plt.axvline(media + 2*desviacion_estandar, color='orange', linestyle='dashed', linewidth=2, label='Desviación Estándar')
plt.axvline(media + 3*desviacion_estandar, color='orange', linestyle='dashed', linewidth=2, label='Desviación Estándar')
plt.axvline(media - desviacion_estandar, color='orange', linestyle='dashed', linewidth=2, label='Desviación Estándar')
plt.axvline(media - 2*desviacion_estandar, color='orange', linestyle='dashed', linewidth=2, label='Desviación Estándar')
plt.axvline(media - 3*desviacion_estandar, color='orange', linestyle='dashed', linewidth=2, label='Desviación Estándar')
# Configuración del gráfico
plt.title('Distribución de Datos con Métricas Estadísticas')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.legend()

# Mostrar el gráfico
plt.show()
```

# **Hipótesis**
* **Hipótesis nula :** la media de la muestra es igual a la media de la población hipotética o conocida
* **Hipótesis alternativa :** la media de la muestra no es igual a la media de la población hipotética o conocida (de dos colas o de dos lados)
* **Hipótesis alternativa :** la media de la muestra es mayor o menor que la media de la población hipotetizada o conocida ( unilateral o unilateral )

# **Fórmula**
$$\it{t} = \frac{ \bar{x} - \mu }{ s / \sqrt{n} }$$

Sigue aproximadamente $t$-distribución con $n-1$ grados de libertad

Dónde,

$X=$ media muestral; \
$\mu=$ media poblacional hipotética o conocida; \
$s=$ desviación estándar de la muestra y \
$n=$ es el tamaño de la muestra

```python
import seaborn as sns
# Crea una figura con 10 subplots (uno para cada columna)
fig, axes = plt.subplots(3, 2, figsize=(10, 10))
fig.subplots_adjust(hspace=0.5)  # Ajusta el espacio entre los subplots

# Itera a través de las columnas y crea los boxplots
for i, column in enumerate(df.select_dtypes(include=np.int64).columns.drop("Restaurant ID")
):
    row = i // 2
    col = i % 2
    sns.boxplot(data=df, y=column, ax=axes[row, col])
    axes[row, col].set_title(column)

# Añade un título general a la figura
fig.suptitle('Boxplots de variables numericas')

# Muestra la figura
plt.show()
```
